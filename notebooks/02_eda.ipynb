```python
# 02_eda.ipynb

# ========================
# Exploratory Data Analysis (EDA)
# ========================

# --- 1. Imports & Setup ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set(style="whitegrid", palette="muted", font_scale=1.1)

# --- 2. Load Processed Data ---
processed_dir = Path("../data/processed")   # adjust path if needed
catalog = pd.read_parquet(processed_dir / "merged_catalog.parquet")
features = pd.read_parquet(processed_dir / "features.parquet")

print("Merged catalog shape:", catalog.shape)
print("Features shape:", features.shape)

# --- 3. Quick Overview ---
display(catalog.head())
catalog.info()
print("\nSource distribution:\n", catalog['source'].value_counts())

# --- 4. Missing Values Analysis ---
missing = catalog.isnull().mean().sort_values(ascending=False)
print("\nTop 15 columns with most missing values:\n", missing.head(15))

plt.figure(figsize=(14,5))
sns.heatmap(catalog.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()

# --- 5. Summary Stats ---
print("\nSummary of numerical features:")
display(features.describe().T)

# --- 6. Distribution of Numerical Features ---
numeric_cols = features.select_dtypes(include='number').columns
features[numeric_cols].hist(figsize=(20, 20), bins=50)
plt.suptitle("Histograms of Numerical Features", y=0.92, fontsize=18)
plt.show()

# --- 7. Correlation Heatmap ---
corr = features[numeric_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr, cmap="coolwarm", center=0, linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()

# --- 8. Target Variable Distribution ---
if "disposition" in catalog.columns:
    print("\nDisposition counts:\n", catalog["disposition"].value_counts())

    plt.figure(figsize=(8,5))
    sns.countplot(data=catalog, x="disposition", order=catalog["disposition"].value_counts().index)
    plt.title("Target Distribution (Disposition)")
    plt.xticks(rotation=45)
    plt.show()

# --- 9. Selected Feature vs Target ---
key_features = ["pl_orbper", "pl_rade", "pl_eqt"]  # Orbital period, radius, equilibrium temp (adjust if missing)
for f in key_features:
    if f in catalog.columns:
        plt.figure(figsize=(8,5))
        sns.boxplot(data=catalog, x="disposition", y=f)
        plt.title(f"{f} vs Disposition")
        plt.xticks(rotation=45)
        plt.show()

# --- 10. Automated Feature Importance (Random Forest, numeric-only) ---

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

if "disposition" in catalog.columns:
    # Prepare X (numeric only) and y
    X = features.select_dtypes(include=["number"]).copy()
    y = catalog["disposition"]

    # Encode target labels
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)

    # Fill missing numeric values
    X = X.fillna(0)

    # Train-test split
    X_train, X_val, y_train, y_val = train_test_split(
        X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
    )

    # Fit Random Forest
    rf = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
    rf.fit(X_train, y_train)

    # Feature importance
    importances = pd.DataFrame({
        "feature": X.columns,
        "importance": rf.feature_importances_
    }).sort_values(by="importance", ascending=False)

    print("\nTop 20 Features by Importance (Random Forest):")
    display(importances.head(20))

    # Plot top 20 features
    plt.figure(figsize=(10,6))
    sns.barplot(x="importance", y="feature", data=importances.head(20), palette="viridis")
    plt.title("Top 20 Important Features (Random Forest)")
    plt.show()
```
